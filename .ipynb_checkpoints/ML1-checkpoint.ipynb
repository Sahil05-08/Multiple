```python
import pandas as pd
d=pd.read_csv("placement.csv")
d
```

```python
import matplotlib.pyplot as tlt
tlt.scatter(d["cgpa"],d["package"])
tlt.xlabel("CGPA")
tlt.ylabel("package")
```

```python
X=d[["cgpa"]]#independent features or i/p
y=d[["package"]]#dependent or o/p
```

```python
from sklearn.model_selection import train_test_split 
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=20)#diveding data into 20 and 80 % in test size
```

```python
X_train
#y_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
print(dir(model))
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
model.predict([[6.89]])
```

```python
y_test
```

```python
y_pred_test=model.predict(X_test)
y_pred_test
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred_test)
```

```python
y_pred_train=model.predict(X_train)
r2_score(y_train,y_pred_train)
```

```python
import joblib

#save the model to the file
joblib.dump(model,'regression_model.joblib')
```

```python
joblib_model=joblib.load('regression_model.joblib')
```

```python
joblib_model.predict([[6.89]])
```
